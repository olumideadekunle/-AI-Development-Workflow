Interactive restore (tune contrast = alpha, brightness = beta)

Use sliders to find the values that best restore the image.

```python
def adjust_image(image_bgr, alpha=1.0, beta=0):
    # alpha: contrast, beta: brightness
    out = cv2.convertScaleAbs(image_bgr, alpha=alpha, beta=beta)
    return cv2.cvtColor(out, cv2.COLOR_BGR2RGB)

@interact(
    alpha=FloatSlider(value=1.0, min=0.1, max=3.0, step=0.05, description='Contrast (alpha)'),
    beta=IntSlider(value=0, min=-100, max=100, step=1, description='Brightness (beta)')
)
def show_restoration(alpha, beta):
    restored_rgb = adjust_image(img_bgr, alpha=alpha, beta=beta)
    plt.figure(figsize=(12,6))

    plt.subplot(1,2,1)
    plt.title("Corrupted")
    plt.imshow(img_rgb)
    plt.axis('off')

    plt.subplot(1,2,2)
    plt.title(f"Restored (alpha={alpha:.2f}, beta={beta})")
    plt.imshow(restored_rgb)
    plt.axis('off')

    plt.show()
```

Save the restored image and record values

```python
# <<< Replace these with your chosen values >>>
best_alpha = 1.35
best_beta = 18

restored_rgb = adjust_image(img_bgr, alpha=best_alpha, beta=best_beta)

plt.figure(figsize=(6,6))
plt.title(f"Final Restored (alpha={best_alpha}, beta={best_beta})")
plt.imshow(restored_rgb)
plt.axis('off')
plt.show()

# Save the restored image
restored_bgr = cv2.convertScaleAbs(img_bgr, alpha=best_alpha, beta=best_beta)
out_name = f"restored_alpha{best_alpha}_beta{best_beta}.jpg"
cv2.imwrite(out_name, cv2.cvtColor(restored_rgb, cv2.COLOR_RGB2BGR))
print("Saved:", out_name)
```

## Question 3: Image Processing  Restoring Corrupted Picture


```python
# Import libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load corrupted image
# Make sure the image is in the same folder as this notebook
# img = cv2.imread("corrupted_picture.jpg")

# Create a dummy image (replace this with your actual image loading)
img = np.zeros((300, 500, 3), dtype=np.uint8)
img[:, :, 0] = 255  # Red channel full
img[:, :, 1] = 100  # Green channel partial
img[:, :, 2] = 50   # Blue channel partial


# Function to adjust brightness & contrast
def adjust_image(image, alpha=1.0, beta=0):
    """
    alpha: Contrast control (1.0-3.0)
    beta: Brightness control (0-100)
    """
    return cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

# Try restoring (adjust alpha & beta until original is restored)
alpha = 1.5   # example value
beta = 30     # example value

restored = adjust_image(img, alpha, beta)

# Show results
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.title("Corrupted Image")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.subplot(1,2,2)
plt.title("Restored Image")
plt.imshow(cv2.cvtColor(restored, cv2.COLOR_BGR2RGB))
plt.show()

print(f"Final Contrast (alpha): {alpha}")
print(f"Final Brightness (beta): {beta}")
```
