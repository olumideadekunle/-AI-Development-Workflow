**The Voice UI Loop — Generative UI Driven by Voice**

In this lesson you'll build a voice agent that drives a tic-tac-toe board on the learner's screen. Voice in. Click in. Same surface.

The architectural piece you'll meet is **Client Actions** — Vocal Bridge's bidirectional channel between the agent and your UI:

- **Agent → App** — voice triggers UI changes (`show_tic_tac_toe`, `place_mark`, …)
- **App → Agent** — clicks and app state flow back as context (`user_move`)

Both directions ride the same WebRTC data channel. No polling. No parsing speech yourself.

````python
import json
import os
from pathlib import Path
import requests
from google.colab import userdata

# Retrieve VOCAL_BRIDGE_API_KEY from Colab secrets.
# If you don't have one, create a key in Vocal Bridge and add it to Colab secrets
# under the name `VOCAL_BRIDGE_API_KEY`.
VOCAL_BRIDGE_API_KEY = userdata.get('VOCAL_BRIDGE_API_KEY')
assert VOCAL_BRIDGE_API_KEY, "Please set VOCAL_BRIDGE_API_KEY in Colab secrets."
print("VB API key loaded ✓")

# Helper class to allow dot-notation access to token response, mimicking the original helper.
class TokenResponse:
    def __init__(self, data):
        self.__dict__ = data

# The mint_token function, as detailed in the notebook's markdown explanation (cell e8f2dcbd).
def mint_token(agent_id, api_key=VOCAL_BRIDGE_API_KEY, participant_name="Notebook Learner"):
    res = requests.post(
        "https://vocalbridgeai.com/api/v1/token",
        headers={
            "X-API-Key": api_key,
            "X-Agent-Id": agent_id,
            "Content-Type": "application/json",
        },
        json={"participant_name": participant_name},
    )
    res.raise_for_status() # Raise an exception for bad status codes
    return TokenResponse(res.json())

# The 'vb' and 'append_to_env' functions will be defined in a separate cell
# after installing the 'vocalbridge' CLI tool.
# The 'voice_widget' function, which renders a React UI component, cannot be
# fully replicated without the original helpers.py file or specific instructions
# for embedding the React component.
````

## Step 2 — Create your Vocal Bridge agent

We'll create a brand-new VB agent from these two files. The CLI does it in one command.

The `--style Chatty` flag picks the OpenAI Realtime backend (good defaults). `--deploy-targets web` keeps it cheap — no phone number provisioned.

**Note**: If this is your first time running the cell below, it will create an `agent_id` for you. If you have already run the cell or other agent provision cell in another notebook, you will see the output as "Using pre-provisioned agent: ..."


````python
import os

# Idempotent: if an agent ID is already set, reuse it. Otherwise create one.
agent_id = os.environ.get("VOCAL_BRIDGE_AGENT_ID_L2", "").strip()
if agent_id:
    print(f"Using pre-provisioned agent: {agent_id}")
else:
    result = vb(
        "agent", "create",
        "--name", "lesson-2-voice-app",
        "--style", "Chatty",
        "--prompt-file", "agents/voice-app/prompt.md",
        "--client-actions-file", "agents/voice-app/actions.json",
        "--deploy-targets", "web",
        json_output=True,
    )
    agent = result["agent"]
    agent_id = agent["id"]
    append_to_env("VOCAL_BRIDGE_AGENT_ID_L2", agent_id)
    print(f"Created '{agent['name']}'  ({agent['mode']})")
    print(f"  id: {agent_id}")
    print(f"  saved to .env as VOCAL_BRIDGE_AGENT_ID_L2")
````

## Step 3 — Mint a voice token

To connect the browser to your VB agent, your backend mints a short-lived **voice session token** by POSTing to `/api/v1/token` with your API key. The browser uses that token to open the session — your API key never leaves the server.

In a real app, the entire backend route is ~10 lines:

```python
import requests

def mint_token(agent_id, api_key):
    res = requests.post(
        "https://vocalbridgeai.com/api/v1/token",
        headers={
            "X-API-Key": api_key,
            "X-Agent-Id": agent_id,
            "Content-Type": "application/json",
        },
        json={"participant_name": "Web User"},
    )
    return res.json()  # → { token, livekit_url, room_name, ... }
```

That's exactly what `mint_token()` in `helpers.py` does.


````python
token = mint_token(agent_id, participant_name="Notebook Learner")
print(f"session_name:   {token.session_name}")
print(f"agent_mode:     {token.agent_mode}")
print(f"token (truncd): {token.token[:40]}…")
# token.connection_url is the wss:// endpoint the SDK opens — kept off-screen to keep this clean.
````

### The pattern that powers all of this

The whole widget is built around **one React hook** — `useAgentActions` — giving you both directions of Client Actions on a single channel:

```jsx
import { useAgentActions } from '@vocalbridgeai/react';

function TicTacToe() {
  const { onAction, sendAction } = useAgentActions();

  // Listen for actions FROM the agent
  useEffect(() => {
    return onAction('show_tic_tac_toe', (payload) => {
      setBoard(Array(9).fill(''));
      setBoardVisible(true);
    });
  }, [onAction]);

  // Send actions TO the agent
  const handleClick = (i) => {
    sendAction('user_placed_mark', {
      position: i, board: nextBoard, status, turn, ...
    });
  };
}
```

Same hook, both directions, on the same WebRTC data channel. Vocal Bridge handles the rest. **This is the architectural piece that's unique to Vocal Bridge** — most voice platforms make you choose between speaking and reacting; here they're one channel.


## Step 5 — Read the call log

When you end the session, VB stores a structured log of every turn and every action. You can pull the most recent one from the CLI:


````python
vb("agent", "use", agent_id)
logs = vb("logs", "--limit", "1", json_output=True)
if not logs:
    print("No calls yet — try the widget above first.")
else:
    print(vb("logs", "show", logs['sessions'][0]['id']))
````
